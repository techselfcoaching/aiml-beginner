```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = {
"Name": ["Alice", "Bob", "Charlie", "David"],
"Age": [24, 30, 22, 35],
"Score": [85, 90, 78, 92]
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
#Simple visualization

df.plot(x='Name', y='Score', kind='bar', title='Scores of Students')
plt.show()
```
